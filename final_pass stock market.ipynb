```python
from statsmodels.tsa.stattools import adfuller
def test_stationarity(timeseries):
    
    #Determing rolling statistics
    rolmean = timeseries.rolling(12).mean()
    rolstd = timeseries.rolling(12).std()

    #Plot rolling statistics:
    orig = plt.plot(timeseries, color='blue',label='Original')
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    std = plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show(block=False)
    
from statsmodels.tsa.stattools import adfuller
def adf_test(timeseries):
    #Perform Dickey-Fuller test:
    print ('Results of Dickey-Fuller Test:')
    test = adfuller(timeseries, autolag='AIC')
    output = pd.Series(test[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in test[4].items():
       output['Critical Value (%s)'%key] = value
    print (output)
```

```python
#apply adf test on the series
test_stationarity(vi_close)
adf_test(vi_close)
```

```python
test_stationarity(ba_close)
adf_test(ba_close)
```

```python
vi_model = ARIMA(vi_train_data, order=(1, 1, 1))  
vi_fitted = vi_model.fit(disp=-1)  
print(vi_fitted.summary())
```

```python
# Forecast
# alpha signifies 95% confidence, 369 is the number of rows in the test dataset taken from the end.
vi_fc, vi_se, vi_conf = vi_fitted.forecast(369, alpha=0.05)  
#converts the fc array into a time series with date as the index 
vi_fc_series = pd.Series(vi_fc, index=vi_test_data.index)
#converts 0th column of the confidence interval array into the upper bound series with time as the index 
vi_lower_series = pd.Series(vi_conf[:, 0], index=vi_test_data.index)
#converts 0th column of the confidence interval array into the upper bound series with time as the index 
vi_upper_series = pd.Series(vi_conf[:, 1], index=vi_test_data.index)
plt.figure(figsize=(12,5), dpi=100)
plt.plot(vi_train_data, label='training')
plt.plot(vi_test_data, color = 'blue', label='Actual Stock Price')
plt.plot(vi_fc_series, color = 'orange',label='Predicted Stock Price')
plt.fill_between(vi_lower_series.index, vi_lower_series, vi_upper_series, 
                 color='k', alpha=.10)
plt.title('Vodafone Idea Stock Price Prediction')
plt.xlabel('Time')
plt.ylabel('Actual Stock Price')
plt.legend(loc='upper left', fontsize=8)
plt.show()
```

```python
ba_model = ARIMA(ba_train_data, order=(0, 1, 0))  
ba_fitted = ba_model.fit(disp=-1)  
print(ba_fitted.summary())
```

```python
# report performance
vi_mse = mean_squared_error(vi_test_data, vi_fc)
print('MSE: '+str(vi_mse))
vi_mae = mean_absolute_error(vi_test_data, vi_fc)
print('MAE: '+str(vi_mae))
vi_rmse = math.sqrt(mean_squared_error(vi_test_data, vi_fc))
print('RMSE: '+str(vi_rmse))
vi_mape = np.mean(np.abs(vi_fc - vi_test_data)/np.abs(vi_test_data))
print('MAPE: '+str(vi_mape))
```
